```python
data=wyniki2015
# Testy normalności rozkładu
from scipy.stats import shapiro, normaltest, chisquare, jarque_bera, kstest
from statsmodels.stats.diagnostic import lilliefors

stat, p =shapiro(data)
print('Shapiro-Wilk Test: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')

stat, p =normaltest(data)
print('D’Agostino’s K-squared test: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')

stat, p =chisquare(data)
print('Chi-Square Normality Test: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')

stat, p =lilliefors(data)
print('Lilliefors Test for Normality: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')

stat, p =kstest(data,'norm')
print('Kolmogorov-Smirnov test for Normality: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')

stat, p =jarque_bera(data)
print('Works when samples >2000 \nJarque–Bera test for Normality: stat=%.3f, p=%3f' %(stat,p)) 
if p>0.05: print('normal\n') 
else: print('not normal\n')
```

#2. Czy wariancje są takie same?

H0:  ro^2_2015 = ro^2_2016

H1: ro^2_2015 != ro^2_2016

Teraz sprawdzmy równość wariancji służy do tego Test F 

**Uwaga !
Napisana poniżej funkcja nie uwzględnia warunku, która z wariancji jest większa.
Nalezy to poprawić/dopisać/sprawdzić.**

image.png


```python
from scipy import stats
# TO WAŻNE, KTÓRA WARIANCJA JEST WIĘKSZA  
y=wyniki2016
x=wyniki2015
print(np.var(x, ddof=1))
print(np.var(y, ddof=1))

f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
dfn = x.size-1 #define degrees of freedom numerator 
dfd = y.size-1 #define degrees of freedom denominator 
p_value = 1-stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 

print('Test F sprawdza czy dwie wariancje populacji są równe: f=%.3f, p_value=%3f' %(f,p_value))

```

Czy wariancje są takie same? Czy uzyskany wynik testu zgadza się z tym, co sądzili Państwo na podstawie wykresu?

Odp: wartość p =0,000944, jest mniejsza od wartości a=0,05, więc powinniśmy odrzucić hipotezę zerową i przyjąć hipotezę alternatywną

#3. Test na równość dwóch średnich
W przypadku tych danych równość średnich sprawdzamy testem t-Studenta (patrz schemat na początku zadania).

H0: średnia wynikó w 2015 jest taka sama jak w 2016

H1: średnia wyników w 2015 jest inna niż w 2016


```python
import scipy.stats as stats

a=wyniki2015
b=wyniki2016
#wykonanie testu t dla dwóch prób z równymi wariancjami, jeśli nie sa równe należy to zmienić
stat, p =stats.ttest_ind(a, b, equal_var=True)
print(stat, p)
```

Czy średnie wyniki w obu rocznikach są takie same?
Odp: na poziomie istotności 0,05 nie mamy podstaw do odrzucenia hipotezy zerowej, że średnie są takie same.
